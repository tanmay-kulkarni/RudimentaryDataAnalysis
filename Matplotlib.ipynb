# Matplotlib Basic Operations

### Line Graph

```python
%matplotlib inline

from scipy.stats import norm
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-5,5,0.05)
plt.plot(x,norm.pdf(x))
plt.show()
```

### I can plot multiple plots on the same graph

```python
plt.plot(x,norm.pdf(x))
plt.plot(x,norm.pdf(x,2,0.5))
plt.show()
```

### Save the created plot to a file

```python

plt.plot(x,norm.pdf(x))
plt.plot(x,norm.pdf(x,2,0.5))
plt.savefig('/home/tanmay/ds/RudimentaryDataAnalysis/linegraph.png',format='png')
```

### But the above graph doesn't tell much. I need to add some information to it to make it convey some meaning to the reader.

```python
import numpy as np
axes = plt.axes()
axes.set_xlim([-5,5])
axes.set_ylim([0,1.0])
axes.set_xticks(range(-5,6))
axes.set_yticks(np.arange(0.0,1.0,0.1))
axes.grid()
plt.plot(x,norm.pdf(x),'r-')
plt.plot(x,norm.pdf(x,2,0.5),'g:')
plt.xlabel('range of numbers')
plt.ylabel('Probability')
plt.legend(['About mean 0','About mean 2'],loc=2)
plt.show()
```

### XKCD style

```python
%config IPCompleter.greedy=True

plt.xkcd()
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
plt.xticks([])
plt.yticks([])
ax.set_ylim([-30, 10])

data = np.ones(100)
data[70:] -= np.arange(30)

plt.annotate(
    'THE DAY I REALIZED\nI COULD COOK BACON\nWHENEVER I WANTED',
    xy=(70, 1), arrowprops=dict(arrowstyle='->'), xytext=(15, -10))

plt.plot(data)

plt.xlabel('time')
plt.ylabel('my overall health')
```

### Pie Chart

```python
# remove xkcd mode
plt.rcdefaults()
values = [12,55,4,32,15]
colors = ['r','g','c','m','b']
explode = [0,0,0,0,0.1]
labels = ['USA','Canada','Pakistan','Ireland','India']
plt.pie(values,colors=colors,labels=labels,explode=explode)
plt.title('Pollution by countries')
plt.show()
```

### Bar Chart

```python
values = [12,33,1,42,53]
colors = ['r','g','b','c','m']
plt.bar(range(1,6),values,color=colors)
plt.show()
```

### Scatter Plot

```python
from pylab import randn
x=randn(500)
y=randn(500)
plt.scatter(x,y)
plt.show()
```

### Histogram

```python
incomes = np.random.normal(65000,30000,10000)
plt.hist(incomes,100)
plt.show()
```
