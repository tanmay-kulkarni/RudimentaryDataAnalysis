# This document demonstrates the significance of Corelation and Covariance. 
### We'll find out these two properties between the gross monthly income of people and the amount of money they spend on entertainment. To that end, let us fabricate some data. First we'll calculate them using the hard way - i.e. by the definitions. Then we'll use the functions readily available in numpy.

#### Formula of Covariance. It indicates how two variables are related. A positive covariance means the variables are positively related, while a negative covariance means the variables are inversely related. 

```python
%%latex
$cov_{x,y}=\frac{\sum_{i=1}^{N}(x_{i}-\bar{x})(y_{i}-\bar{y})}{N-1}$
```

```python
%matplotlib inline
import numpy as np
from pylab import *

def deviation_from_mean(X):
    xmean = mean(X)
    return [xi - xmean for xi in X] # returns a list of mean deviations
    
def covariance(x,y):
    n = len(x)
    return dot(deviation_from_mean(x),deviation_from_mean(y))/n-1

incomes = np.random.normal(80000,20000,1000)
money_spent = np.random.normal(30000,5000,1000)

scatter(money_spent,incomes)
covariance(money_spent,incomes)
```

#### Let's now make a meaningful relationship between the two variables.

```python
money_spent = np.random.normal(5000,1000,1000)+incomes/5
scatter(money_spent, incomes)
covariance(money_spent,incomes)
```

#### Correlation is another way to determine how two variables are related. In addition to telling you whether variables are positively or inversely related, correlation also tells you the degree to which the variables tend to move together.

```python
%%latex
$\rho = \frac{\text{cov}(X,Y)}{\sigma_x \sigma_y}$
```

```python
def corelation(x,y):
    std_x = x.std()
    std_y = y.std()
    return covariance(x,y)/(std_x*std_y)

corelation(money_spent,incomes)
```

#### Numpy provides np.corrcoeff() and np.cov() to calculate these directly.

```python
np.corrcoef(money_spent,incomes)
```

```python
np.cov(money_spent,incomes)
```
