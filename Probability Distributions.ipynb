# Here are some important distributions: 

## 1. Uniform Distribution

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

values = np.random.uniform(-100,100,5000)
plt.hist(values,100)
plt.show()
```

## 2. Normal or Gaussian distribution

```python
from scipy.stats import norm
import matplotlib.pyplot as plt

x = np.arange(-5,5,0.01)
plt.plot(x, norm.pdf(x))
```

## Generate some random normally distributed data using NumPy

```python
import numpy as np
import matplotlib.pyplot as plt

mu = 10.0
sigma = 3.0
data  = np.random.normal(mu,sigma,10000)

plt.hist(data,50)
plt.show()
```

## 3. Exponential Probability Distribution Function

```python
from scipy.stats import expon
import matplotlib.pyplot as plt

x = np.arange(0,10,0.001)
plt.plot(x,expon.pdf(x))
```

## 5. Poisson Distribution. 
A website gets 500 visits per day. What is the probability of getting 575?

```python
from scipy.stats import poisson
import matplotlib.pyplot as plt
mu = 500
x = np.arange(400,600,0.5)
plt.plot(x,poisson.pmf(x,mu))
```
